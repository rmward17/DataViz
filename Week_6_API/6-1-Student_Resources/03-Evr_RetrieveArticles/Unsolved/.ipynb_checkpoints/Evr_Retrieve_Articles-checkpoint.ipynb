```python
# Dependencies
import requests
from config import api_key

url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Store a search term
search_term = "basketball"

# Search for articles published between a begin and end date
begin_date = "20200101"
end_date = "20222903"
# Build url
nyt_url = url +"&fq=" + search_term + "&api-key=" + api_key
```

```python
# Retrieve articles
article = requests.get(nyt_url)
```

```python
article = article.json()
```

```python
articles_list = [article for articles in article["response"]["docs"]]
print(articles_list[0])
```
