```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV
got_data_df = pd.read_csv("Resources/got.csv")
got_data_df.head()
```

```python
# Get attacker and defender king data
attacker = got_data_df.groupby("attacker_king").value_count()
defender = got_data_df.groupby("defender_king").value_count()
```

```python
# Get total battle data
total_battles = pd.DataFrame({"Attacker":attacker, "Defender":defender}).fillna(0)
total_battles
```

```python
# Configure plot and ticks
total_battles.plot(kind="bar")

# Set textual properties


# Show plot


# Resize plot to display labels
```
