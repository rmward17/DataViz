```python
import numpy as np
import matplotlib.pyplot as plt

x_axis = np.arange(len(rain_df))

tick_locations = [value for value in x_axis]

plt.figure(figsize=(20,3))
plt.bar(rain_df["State"],rain_df["Inches"])
plt.xticks(tick_locations,rain_df["State"], rotation='vertical')

plt.title("Rain in Inches by State")
plt.xlabel("State")
plt.ylabel("Rain (inches)")
plt.show()
```

```python
test_grades = {
    'Class': ['Oct', 'Oct', 'Jan', 'Jan', 'Oct', 'Jan'], 
    'Name': ["Cyndy", "Logan", "Laci", "Elmer", "Crystle", "Emmie"], 
    'Test Score': [90, 59, 72, 88, 98, 60]}
grades_df = pd.DataFrame(test_grades)
grades_df.head()
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 59,69,79,89,100]

# Create labels for these bins
group_labels = ["F","D","C","B","A"]

# Slice the data and place it into bins
grades_df["Letter Grades"] = pd.cut(grades_df["Test Score"], bins, labels=group_labels)
grades_df.head()
```
